### b) - Använd den tidigare ratings per film sorteringen för att räkna medelrating per film

```python
#Definierar data variabler som innehåller inläst data från csv formaterade filer
ratingData2 = pd.read_csv("./u.data", sep="\t", names=["usersId", "itemId", "rating", "timespamp"])
```

```python
#Sorterar enligt ratings och nollar indexen
ratingData2.sort_values(by = 'rating', ascending = False, ignore_index = True)

#Grupperar enligt itemId och gör indexen från 0, samt tar medelvärdet
ratingData3 = ratingData2.groupby('itemId', as_index=False).mean()
ratingData3[["itemId", "rating"]]
```

### c) - Hitta top 10 filmerna bland de som har fler än 40 ratings

```python
#Väljer endast top 10 med head(10)
ratingTop10.sort_values(ascending = False).head(10)
```

## a) - Använd u.user datan för att undersöka vilka yrken som är vanliga bland män och viceversa?

```python
#Definierar data variabler som innehåller inläst data från csv formaterade filer
#Maneullt jämför yrken
userData5 = pd.read_csv("./u.user", sep="|", encoding="latin1", names = ["Index","Age","Gender", "Profession", "?"])
userData6 = userData5[(userData5['Gender'].str.startswith('M'))]
userData6[userData6.Profession == "technician"]
```

Det är betydligt mycket mer allmänt för män att vara ingenjörer än kvinnor enligt datan

```python
userData6 = userData5[(userData5['Gender'].str.startswith('F'))]
userData6[userData6.Profession == "technician"]
```

26:1 för ingenjör yrket

```python
#Gruppering på rad 1, räkneoperation på rad 2 (kollar skillnaden), rad 3 visar dominans i ett yrke
userDataComparisons = userData5.groupby(['Profession','Gender']).size().reset_index(name='counts').pivot(index='Profession', columns='Gender').fillna(0)
userDataComparisons['Delta'] = userDataComparisons['counts', 'M'] - userDataComparisons['counts', 'F']
userDataComparisons['Domination'] = np.where(userDataComparisons['Delta'] > 0, "Male","Female")
userDataComparisons
```

## b) - Kombinera u.data med u.item för att  skapa en tabell där man kan se Movie Title och Rating

### c) - Kombinera sedan den datan med datan från användartabellen (u.user) för att visafilmtitel, rating och information om användaren som gett den ratingen.

### d) - Vilka 5 användare är de snällaste och vilka 5 användare är de tuffaste? (Slå ihop varjeanvändares filmratings till ett medeltal)
